### Grouping and aggregation

To summarize the daywise data and create a new dataframe with month-wise data. This is where the groupby funtion is useful. Along with a grouping, we need to
specify a way to aggregate the data for each group.

```python
covid_month_df = covid_df.groupby('month')[['new_cases', 'new_tests','new_deaths']].sum()
```

```python
covid_month_df
```

```python
covid_monthgroups_df = (
    covid_df.groupby('month')[['new_cases', 'new_tests', 'new_deaths']]
    .sum()
    .sort_values(by='new_cases', ascending=False)
)
```

```python
covid_monthgroups_df
```

```python
covid_month_mean_df= covid_df.groupby('weekday')[['new_cases', 'new_tests','new_deaths']].mean()
```

```python
covid_month_mean_df
```

Another form of aggregation is to calculate the running or cumulative sum of
cases, tests or death up to the current date for each row. This can be done using cumsum
method. Let's add 3 new columns: total_cases , total_deaths and total_tests .

```python
covid_df['total_cases'] = covid_df.new_cases.cumsum()

covid_df['total_deaths' ] = covid_df.new_deaths.cumsum()

covid_df['total_tests'] = covid_df.new_tests.cumsum() + initial_tests
```

```python
covid_df
```

### Merging data from multiple sources

To determine other metrics like test per million, cases per million etc. we require more some
information about the country viz. it's population. Let's download another file locations.csv
which contains health-related information for different countries around the world, including Italy.

```python
from urllib.request import urlretrieve
```

```python
import requests
```

```python
# GitHub mirror (same data)
url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"

# Download and save locally
r = requests.get(url)
open("owid-covid-data.csv", "wb").write(r.content)

# Load into pandas
df1 = pd.read_csv("owid-covid-data.csv")
```

```python
# Extract key metadata
locations_df = df1[['location', 'continent', 'population', 'life_expectancy',
                   'hospital_beds_per_thousand', 'gdp_per_capita']].drop_duplicates(subset='location')

# Save subset
locations_df.to_csv("locations.csv", index=False)
```

```python
locations_df
```

```python
locations_df[locations_df['location']== 'Italy']
```

To merge this data into our existing data frame by adding more columns. However, to merge
two data frames, we need at least one common column. So let's insert a location column in
the covid df dataframe with all values set to "Italy"

```python
# add column location with all values set to Italy
covid_df['location']='Italy'
```

```python
covid_df
```

##### add the columns from locations_df into covid_df using the .merge
method.

```python
merged_df= covid_df.merge(locations_df, on='location')
```

```python
merged_df
```

```python
merged_df['location'].unique()
```

```python
import math
```

```python
merged_df['cases_per_million'] = merged_df.total_cases * 1e6 / merged_df.population
merged_df['deaths_per_million'] = merged_df.total_deaths * 1e6 / merged_df.population
merged_df['tests_per_million'] = merged_df.total_tests * 1e6 / merged_df.population
```

```python
merged_df
```
