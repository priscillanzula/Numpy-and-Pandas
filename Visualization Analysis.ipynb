### Plotting Multiple Lines
It's really easy to plot multiple lines in the same graph. Just invoke the plt.plot function
multiple times. Let's compare the yields of apples vs. oranges in Kanto.



```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.97375,0.9372,0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898,0.9,0.896]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('years')
plt.ylabel('Yield() tons per hectare')
```

### Chart Title and legend

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('years')
plt.ylabel('Yield() tons per hectare')

plt.title('Yield per crop')
plt.legend(['oranges','apples']);
```

### Line markers


```python
plt.plot(years, apples, marker= 'x')
plt.plot(years, oranges, marker='*')
plt.xlabel('years')
plt.ylabel('Yield() tons per hectare')

plt.title('Yield per crop')
plt.legend(['oranges','apples']);
```

### styling lines and markers


##### The plt.plot function supports many arguments for styling lines and markers:

- color or c : set the color of the line .
- linestyle or 1s : choose between a solid or dashed line.
- linewidth or lw : set the width of a line.
- markersize or ms : set the size of markers
- markeredgecolor or mec : set the edge color for markers.
- markeredgewidth or mew : set the edge width for markers.
- markerfacecolor or mfc : set the fill color for markers.
- alpha : opacity of the plot.

In Matplotlib (and many other graphics libraries), the alpha parameter controls transparency  or in other words, how opaque an element is.

###### Heres how it works:

- alpha = 1.0  fully opaque (completely solid color)

- alpha = 0.5  semi-transparent (50% visible, 50% see-through)

- alpha = 0.0  fully transparent (invisible)

```python
plt.plot(years, apples, marker= 'x', c='b', ls='dashed', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='*', c='r', ls='dotted', lw=3, ms=10, alpha= 0.5)
plt.xlabel('years')
plt.ylabel('Yield() tons per hectare')

plt.title('Yield per crop')
plt.legend(['oranges','apples']);
```

#### The fmt argument provides a shorthand for specifying the line style, marker and line color. It can
be provided as the third argument to plt. plot .
- fmt = '[marker ] [ line ] [ color]'

```python
plt.plot(years, apples,'s-b')
# fmt= square-marker, solid- linestyle, blue -line color
plt.plot(years, oranges,'o--r')
# fmt= circle-marker, dashed- linestyle, red -line color
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges' ]) ;
```

#### Adding Hues

Notice how the points in the above plot seem to form distinct clusters with some outliers. We can
color the dots using the flower species as a hue . We can also make the points larger using the
s argument.

```python
sns.scatterplot(x=flowers_df.sepal_length,
y= flowers_df.sepal_width,
hue=flowers_df.species, s=100);
```

- Setosa has the smallest sepal length and  a higher sepal width, while this is opposite for viginica where it has long sepal length and a high sepal width.

### Customizing Seaborn Figures

```python

# Since Seaborn uses Matplotlib's plotting functions internally, we can use functions like
# plt.figure and plt.title to modify the figure.

plt.figure(figsize=(12,6))
plt.title('Sepal Dimensions')

sns.scatterplot (x= flowers_df.sepal_length,
y= flowers_df.sepal_width,
hue=flowers_df.species,
s=100);
```

### Multiple Histograms



Similar to line charts, we can draw multiple histograms in a single chart. We can reduce the opacity
of each histogram, so the the bars of one histogram don't hide the bars for others.





```python
setosa_df = flowers_df [flowers_df['species']== 'setosa' ]
versicolor_df = flowers_df[flowers_df['species'] == 'versicolor' ]
virginica_df = flowers_df[flowers_df.species == 'virginica' ]
```

```python
setosa_df
```

```python
plt.hist(setosa_df['sepal_width'], alpha=0.5, bins=np.arange(2,5,0.25));
plt.hist(virginica_df['sepal_width'], alpha=0.5, bins=np.arange(2,5,0.25));
plt.hist(versicolor_df['sepal_width'], alpha=0.5, bins=np.arange(2,5,0.25));
plt.legend(['virginica', 'versicolor','setosa']);
```

### Bar Chart

```python
years = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
```

```python
# Like histograms, bars can also be stacked on top of one another. We use the bottom argument
# to plt.bar to achieve this.

plt.bar(years, apples)
plt.bar(years, oranges, bottom=apples);
plt.legend(['aplples','oranges'])
```
